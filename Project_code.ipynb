```python
#Representation and Inclusion Efforts Analysis
#Cross-tabulate demographic factors with representation metrics
representation_cross_tab = pd.crosstab(data['Gender'], data['Ethnicity'])    
representation_cross_tab.plot(kind='bar', stacked=True, figsize=(10, 6), title='Representation by Gender and Ethnicity')
```

```python
#Representation and Inclusion Efforts Analysis
# Cross-tabulate demographic factors with representation metrics
representation_cross_tab = pd.crosstab(data['Gender'], data['Ethnicity'], normalize='index')

# Print cross-tabulation
print("Representation by Gender and Ethnicity:")
print(representation_cross_tab)
```

```python
# Leadership Representation Analysis

# Filter data for employees in leadership roles within the IT sector
leadership_data = data[(data['Division'] == 'IT') & (data['Manager'] == 'Yes')]

# Calculate representation of different demographic groups in leadership roles
leadership_representation = leadership_data.groupby(['Gender', 'Ethnicity']).size().unstack(fill_value=0)

# Print representation of different demographic groups in leadership roles
print("Representation of Different Demographic Groups in Leadership Roles within the IT Sector:")
print(leadership_representation)

#visual representation
plt.figure(figsize=(10, 6))
sns.heatmap(data=leadership_representation, annot=True, cmap='YlOrRd')  # Change the colormap here
plt.title('Representation of Different Demographic Groups in Leadership Roles within the IT Sector')
plt.xlabel('Ethnicity')
plt.ylabel('Gender')
plt.show()
```
